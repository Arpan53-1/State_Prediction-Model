### 1. Mutual Information (Information Gain)

Mutual Information measures how much information one variable gives about another. In the context of classification, it measures the dependency between a feature and the target variable. A higher mutual information score indicates that the feature is more significant in predicting the target.

Steps:
1. Calculate Mutual Information between each feature and the target variable.
2. Rank the features based on their mutual information scores.

```python
# Select only the numerical features for calculating mutual information
X_new = X.select_dtypes(include=['float64', 'int64'])
X_new
```

```python
##---------Type the code below this line------------------##
from sklearn.feature_selection import mutual_info_classif
import pandas as pd

# Calculate mutual information between each feature and the target
mi_scores = mutual_info_classif(X_new, y, random_state=42)

# Create a dataframe to display scores
mi_df = pd.DataFrame({'Feature': X_new.columns, 'Mutual Information': mi_scores})
mi_df = mi_df.sort_values(by='Mutual Information', ascending=False)

# Display the top 5 significant features based on mutual information
print(mi_df.head(5))
```

### 2. Chi-Squared Test
The Chi-Squared test evaluates the independence between categorical variables and the target. It measures whether the observed frequency distribution differs from the expected distribution.

Steps:
Apply the Chi-Squared test between categorical or discretized numerical features and the target.
Rank features based on their chi-squared statistic.
Instead of standardizing the data (which can result in negative values), you can normalize the data using MinMaxScaler to ensure all values are between 0 and 1, which is appropriate for the Chi-Squared test.

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import chi2
import pandas as pd

# Normalize numerical features to non-negative values
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_new)

# Perform Chi-Squared test
chi2_scores, p_values = chi2(X_scaled, y)

# Create a dataframe to display Chi-Squared scores and p-values
chi2_df = pd.DataFrame({'Feature': X_new.columns, 'Chi-Squared Score': chi2_scores, 'p-value': p_values})
chi2_df = chi2_df.sort_values(by='Chi-Squared Score', ascending=False)

# Display the top 5 significant features based on Chi-Squared test
print(chi2_df.head(5))
```

### 1. Classification: Decision Tree

```python
##---------Type the code below this line------------------##
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score


# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.3, random_state=42)

# Initialize and train the Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Evaluate the model
print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

## 6.2 ML technique 2 + Justification

### 2. Clustering: K-Means

```python
##---------Type the code below this line------------------##
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt



# Determine the optimal number of clusters using the Elbow Method
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_new)
    inertia.append(kmeans.inertia_)

# Plot the Elbow curve
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()

# Apply K-Means with the optimal number of clusters (assume k=3 for this example)
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(X_new)

# Add cluster labels to the original dataframe
df['Cluster'] = clusters

# Visualize clusters (assuming two principal components for simplicity)
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=df['Cluster'], cmap='viridis')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
